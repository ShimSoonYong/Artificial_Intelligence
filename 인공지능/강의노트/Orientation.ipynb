## 기계학습

최근 인공지능 발전의 주 동력원이 되는 기술이다.  
기계 스스로 대량의 데이터에서 지식이나 패턴을 찾아 학습하여, 새로운 입력이나 상황에 대해 출력을 예측하는 시스템을 달성하는 기술이다.

__Arthur Samuel__:  
_명시적으로 프로그램을 작성하지 않고 컴퓨터에 학습할 수 있는 능력을 부여하기 위한 연구 분야_

__인공지능의 발전 방향__  
- 연결주의(Connectionism)
    - 지능은 복잡하게 연결된 신경망에 의해 이루어진다.
    - 다층 퍼셉트론은 비선형 분리도 가능하다.
- 기호주의(Symbolism)
    - 1970년대 중반 ~ 1980년대 후반까지 인공지능의 핵심 접근법이었다.
    - 기호를 통해 개념을 정의하고, 이 기호들을 가지고 일정한 논리적 규칙에 따라 추론한다.
    - 전문가 시스템으로 발전했다.

__기계학습의 현대적 재정의__  
- Tom Mitchell(Carnegie Mellon University, 1997)
    - 만약 캄퓨터 프로그램이 특정한 작업 T를 수행할 때 성능이 P만큼 개선도ㅚ는 경험 E를 보이면, 그 컴퓨터 프로그램은 작업 T와 성능 P에 대해 경험 E를 학습했다고 할 수 있다.

### 기계학습의 분류

|유형|작업|알고리즘|
|---|---|---|
|지도학습(Supervised Learning)|분류(Classification), 예측(Prediction)|KNN, SVM, Decision Tree, Logistic Regression, Linear Regression|
|비지도학습(Unsupervised Learning)|군집(Clustering)|K-means Clustering, DBSCAN Clustering, Hierarchical Clustering|
|강화학습(Reinforcement Learning)|로봇, 게임 등에 활용|MDP: Markov Decision Process|
