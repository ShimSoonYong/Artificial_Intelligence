# 신경망 기초

__기계학습의 목적__  
$x \to f \to f(x)$  

기본적으로 문제를 해결할 수 있는 함수 $f$를 구축하는 것이다.  
함수의 출력값과 목표 변수(target)의 차이(손실, Loss)가 가능한 한 작아야 한다.  
목표 변수가 이미 주어진 데이터셋을 사용해 기계를 학습시킬 경우 그 데이터의 각 행들을 training example이라고 한다.  
  
그 함수는 훈련 데이터에는 없는 데이터를 입력받아도 적절히 반응하도록(손실이 작도록) 해야 한다. 훈련 데이터에 없는 데이터에 대해서 잘 반응하는 정도를 일반화 성능(Generalization Capability)이라고 한다.

__기계학습 문제__  
훈련 데이터로 함수(시스템, 또는 모델) $f$를 훈련시키고자 한다.  
훈련 데이터는 다수의 훈련 예제(train example)로 이뤄져 있다.  
하나의 훈련 예제는 $<X, d>$와 같은 형태이다. 여기서
- $X$는 특성 벡터 $<x_1, x_2, ..., x_n>$
- $d$는 레이블(또는 target, ground truth, 즉 정답)로, 정답 클래스나 입력된 $X$에 따른 행동을 나타낸다.
  
학습 문제는 결국 함수 $f(x)$를 찾는 것이므로:  
- $f$의 출력은 훈련 예제에 부합한다. 
    - 예를 들어, 훈련 예제 $<X, d>$에 대해 $f(x)$는 $d$를 나타내는 값을 의미한다.

## 지도학습

__훈련 데이터셋__  
- $[1] = {<X_1, d_1>, <X_2, d_2>, ..., <X_n, d_n>}$
    - $<X_i, d_i>$: 훈련 예제
    - $X_i$: 입력 데이터
    - $d_i$: 타겟 레이블, 클래스 레이블, 값(value)
  
__학습의 목표는 $f$를 찾는 것이다.__  
- 그리하여 $[1]$에 있는 최대한 많은 $i$에서 $f(X_i)$가 $d_i$를 나타내도록 하는 것이다.

__이를 지도학습이라고 한다.__ 왜냐하면:  
- 기계가 타겟 레이블 $d_i$를 가진 훈련 예제로부터 학습했기 때문이다.

## 뉴런(Neuron)

신경망 방법론에서, 함수 $f$는 뉴런의 네트워크로 구현된다.  
- 뉴런이 하나만 있으면 $f$는 한 뉴런으로 구성된 것이다. 
- 그때, 파라미터 셋(parameter set), $w_i$와 편향$b$가 $f$를 구성한다.

__뉴런__   
- 신경망 모델에서 사용하는 가장 원시적인 연산 요소이다.
- 뉴런이라고도 불리지만 퍼셉트론, 유닛이라고도 불린다. 
- 학습 모델을 구축하기 위해선 하나 이상의 뉴런이 필요하다.

__뉴런의 구조__  


__뉴런의 활성화 함수(Activation Function)__:  
- 이는 가중합 $s$에 적용된다.
- 다양한 함수들이 활성화 함수로 사용된다.
    - 선형 함수: $a(s)=s$
    - 시그모이드(로지스틱) 함수: $a(s)=sigmoid(s)=\sigma(s)$
    - 하이퍼볼릭 탄젠트 함수: $a(s)=\tanh(s)$
    - Rectified Linear Function(ReLU 함수): $a(s)=ReLU(s)=max(0,s)$

__모델의 파라미터__  
- 단일 뉴런 모델의 파라미터들:
    - $f(X|w_1, w_2, ..., w_n, b)$
    - 뉴련의 가중치와 편향이 파라미터 셋을 구성한다.
        - 이들은 훈련 시에 갱신(update)된다.
    - 이 파라미터들의 특정한 값이 모델을 구축한다.
        - 그들의 값이 결정되는 순간 모델(또는 시스템)이 완전히 정의된다.

## 신경망의 구조

단일 뉴런으로 이루어진 시스템을 <font color='red'>퍼셉트론</font>이라고 한다.  
- 선형 분리 가능한 훈련 셋에서 잘 작동한다.
  
- 훈련 셋이 선형 분리가 불가능하면:
    - 모든 훈련 예제에 대해 올바른 출력을 내는 뉴런을 구축할 수 없다.
    - 단일 뉴런 시스템은 모델로서 너무 약하다.
- 해결책:
    - 시스템 구축에 다수의 뉴런을 사용한다. 이를 Neural Network라고 한다.

__Neural Network(NN)__ 의 구조:
- feed forward 형태이다.
    - cycle이 없어야 한다.
- layer들로 이뤄져 있다.
    - 단일 레이어는 많은 뉴런을 가진다.
- layer 간의 연결은 완전연결(fully-connected)이다.
    - 즉 두 층간의 모든 뉴런은 모두 연결되어 있다.
- 이런 것을 계층화된 완전연결 앞먹임 신경망(layered fully-connected feed-forward neural network)이라고 한다.
    - FC(FFNN), or FCNN
- '다층 퍼셉트론(Multi Layered Perceptron)'이라고도 한다.(<font color='lightblue'>MLP</font>)

```python
import matplotlib.pyplot as plt
import networkx as nx

# 4계층 신경망 시각화를 위해 그래프 생성
G = nx.DiGraph()

# 각 계층의 노드 수 설정
layers = [3, 5, 4, 2]  # 입력, 은닉, 은닉, 출력 계층

# 노드 추가
positions = {}
node_idx = 0
layer_pos = 1
for layer_num, num_nodes in enumerate(layers):
    for i in range(num_nodes):
        node_name = f"L{layer_num+1}N{i+1}"
        G.add_node(node_name)
        positions[node_name] = (layer_pos, i)
    layer_pos += 1

# 엣지 추가 (모든 노드에서 다음 계층의 모든 노드로 연결)
prev_layer_nodes = []
for layer_num, num_nodes in enumerate(layers):
    current_layer_nodes = [f"L{layer_num+1}N{i+1}" for i in range(num_nodes)]
    
    if prev_layer_nodes:
        for prev_node in prev_layer_nodes:
            for curr_node in current_layer_nodes:
                G.add_edge(prev_node, curr_node)
    
    prev_layer_nodes = current_layer_nodes

# 그래프 시각화
plt.figure(figsize=(8, 6))
nx.draw(G, pos=positions, with_labels=True, node_color='skyblue', node_size=2000, arrows=False)
plt.title("4-Layer Neural Network")
plt.show()
```

### NN을 위한 신호(Signals)와 변수(variables)들

$X^{(i)}$: $(i+1)^{th}$번째 계층의 입력 벡터
- $X^{(i)}=(f_1^{(i)}, f_2^{(i)}, ...,f_{m_i}^{(i)})$: 이것은 $i$번째 계층의 출력층에서 받는다.
- $X^{(0)}$: 첫번째 계층의 입력
    - $X^{(0)}$는 훈련 예제$(X,d)$의 $X$에서 받는다.
  
$N_i^{(j)}$: j번째 계층의 i-번째 뉴런을 나타낸다.  
$s_i^{(j)}$: $N_i^{(j)}$에서 온 신호의 가중합이다.  
$f_i^{(j)}$: $N_i^{(j)}$ 뉴런의 출력이다.
  
$\delta_i^{(j)}$: 훈련 도중의 $N_i^{(j)}$ 안의 값이다.  
$m_j$: j번째 계층의 뉴런의 개수이다.
  
$W_i^{(j)}$: $N_i^{(j)}$에 입력되는 신호의 가중치 벡터이다.
- $W_i^{(j)}=(w_{1,i}^{(j)}, ...,w_{m_{j-1},i}^{(j)})^T$
  
$W^{(j)}$는 2차원 행렬이다.
- i번째 열은 $W_i^{(j)}$이다.
- 각 행은 이전 층의 노드에서 나온 각 출력값(가중합 전의 것)에 곱해질 가중치 벡터이다.

## Forward pass of NN: 입력에서 모델의 출력 구하기

$T$ 를 층의 개수라고 하자.

출력층에선 $f_k^0 = x_k^{(0)}$이라고 가정하고, $a^j$를 $j$층에서 사용된 활성화 함수라고 하자.

__알고리즘__  
For $j=1$ to $T$ do  
> For $i=1$ to $m_j$ do {  
        >> $s_i^j = b_i^j + \sum_{k=1}^{m_{j-1}}{f_k^{j-1}w_{k,i}^j}$  
        >> $f_i^j=a^j(s_i^j)$  
> }

마지막 층에 있는 뉴런들의 출력은:  
$(f_1^T,\ ...,\ f_{m_T}^T)$

## 훈련을 위한 경사하강법(Gradient Descent method)

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# 3차원 함수 정의 (예시로 이차 함수 선택)
def f(x, y):
    return x + y

# 함수의 gradient 정의
def grad_f(x, y):
    df_dx = 2 * x
    df_dy = 2 * y
    return np.array([df_dx, df_dy])

# 경사 하강법 함수
def gradient_descent(starting_point, learning_rate, num_iterations):
    path = []
    point = np.array(starting_point)
    for _ in range(num_iterations):
        path.append(point.copy())
        gradient = grad_f(point[0], point[1])
        point = point - learning_rate * gradient
    return np.array(path)

# 시작점, 학습률, 반복 횟수 설정
starting_point = [5, 5]
learning_rate = 0.1
num_iterations = 50

# 경사 하강법 실행
path = gradient_descent(starting_point, learning_rate, num_iterations)

# 3D 플롯 그리기
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# X, Y 범위 설정
x = np.linspace(-6, 6, 400)
y = np.linspace(-6, 6, 400)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

# 함수의 3D 곡면 그리기
ax.plot_surface(X, Y, Z, cmap='viridis', alpha=0.2, rstride=100, cstride=100)

# 경사 하강법 경로 그리기
Z_path = f(path[:, 0], path[:, 1])
ax.plot(path[:, 0], path[:, 1], Z_path, color='r', 
        marker='o', markersize=5, label='Gradient Descent Path')

# 축 레이블 설정
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')
ax.set_title('3D Gradient Descent Visualization')

plt.show()
```

훈련되지 않은 신경망의 가중치와 편향은 무작위 값을 가지고 있어서 엉터리 값들이 출력된다. 따라서 가중치와 편향을 보다 적합한 값으로 조정할 필요가 있다. 이 과정을 학습 또는 훈련이라고 한다.

손실은 $L(X,Z;\theta)$로 표기한다.
- $\theta$는 신경망의 모든 뉴런의 모든 가중치와 편향을 의미한다.
- 손실(Loss)는 파라미터 $\theta$의 함수이다.

우리는 손실이 최소화되는 방법으로 파라미터 $\theta$를 조정한다.
- 이것이 훈련(학습)의 목표이다.
- 따라서 이것은 최적화 문제 중에서 최소화에 해당한다.

경사하강법은 최적화 문제에 사용된다.
- 이를 위해선 파라미터의 각 원소에 대한 그레디언트(기울기)를 계산할 것이 요구된다.
- 파라미터를 갱신하기 위한 규칙은 아래와 같다:
$$
w=w-\eta \frac{\partial L}{\partial w}
$$
여기서:  
$$
\frac{\partial L}{\partial w} = \text{gradient} \\
const.\ \eta\ < 1
$$

모든 훈련 예제를 반복적으로 사용하여 손실이 최소화되는 값이 되도록 모든 파라미터가 갱신된다.  
이 때 각 반복 횟수를 epoch(에포크)라고 한다.  
$\eta$(에타)는 학습률(learning rate)이라고 한다. eta가 너무 크면 아래와 같이 발산해버린다.

```python
import numpy as np
import matplotlib.pyplot as plt

# 가상의 손실 함수: y = x^2
def loss_function(x):
    return x**2

# 손실 함수의 미분: f'(x) = 2x
def gradient(x):
    return 2 * x

# 경사하강법 구현
def gradient_descent(initial_x, learning_rate, num_iterations):
    x_values = [initial_x]
    for _ in range(num_iterations):
        grad = gradient(initial_x)
        initial_x -= learning_rate * grad
        x_values.append(initial_x)
    return x_values

# 초기 값과 학습률 설정
initial_x = 5
learning_rate = 1.1  # 발산을 유도하는 큰 학습률
num_iterations = 10

# 경사하강법 실행
x_values = gradient_descent(initial_x, learning_rate, num_iterations)
loss_values = [loss_function(x) for x in x_values]

# 시각화
fig, ax = plt.subplots()

# x 범위
x_range = np.linspace(-6, 6, 100)
ax.plot(x_range, loss_function(x_range), label="Loss function (x^2)", color='blue')

# 경사하강법의 경로 시각화
ax.plot(x_values, loss_values, 'ro-', label='Gradient Descent Path', markersize=8)

# 그래프 꾸미기
ax.set_title("Divergence with Large Learning Rate in Gradient Descent")
ax.set_xlabel("x")
ax.set_ylabel("Loss (x^2)")
ax.legend()

# 시각화 출력
plt.show()
```

### 그레디언트 계산: 미분의 연쇄법칙(chain-rule) 적용

임의의 파라미터 $w$를 위해 계산될 그레디언트는:
- $\frac{\partial L}{\partial w}$

경우의 수는 두 개:
- case1: $w$가 직접적인 영향을 미치기 때문에 함수 $L$이 변수 $w$을 사용한 수식으로 쓰일 수 있다.
- case2: 아니면 $L$이 $w$에 의존하지만 $w$를 직접 사용하여 표현될 수는 없을 수도 있다.

Case1: 미분을 하면 그레디언트가 바로 나온다.  
Case2: 연쇄법칙을 사용해야 한다.

손실함수 $L$이 $u_1, ...,u_n$이 주어지면 유일하게 결정될 수 있다고 하자.
- 이는 $L$이 이 변수들의 <font color='red'>완전 함수</font>임을 의미한다.
- $L(u_1, ..., u_n)$
- 이제 $\frac{\partial L}{\partial u_i}$가 모든 $i$에 대해 알려져 있다고 하자.

그리고 $w$를 $L$에 영향을 미치는 변수라고 하자.
- 달리 말해, $w$가 변하면 $L$도 변한다.
- 그러나, $w$는 다른 변수들을 통하여 $L$에 영향을 미친다.
    - 예를 들어, 각각의 $u_i$가 $w$에 영향을 받는다.

그러면 $w$의 그레디언트는 연쇄법칙에 따라 아래와 같이 쓰일 수 있다:
$$
\frac{\partial L}{\partial w}=\frac{\partial L}{\partial u_1}\frac{\partial u_1}{\partial w}+ \cdots + \frac{\partial L}{\partial u_n}\frac{\partial u_n}{\partial w}
$$

- 각 $u_i$는 $w$에 직접 영향을 받으므로 바로 미분 가능하다. 즉 각 $i$에 대해 $\frac{\partial u_i}{\partial w}$를 구할 수 있다.
- 이들을 위 식에 대입하면 결국 $\frac{\partial L}{\partial w}$를 구할 수 있다.

#### 연쇄법칙의 반복 적용
$w$가 $L$에서 멀다면, 연쇄법칙은 여러 번 적용되어야 할 수도 있다.

원래 주어진 목표는 $\frac{\partial L}{\partial w}$를 구하는 것이라고 하자.  
$w$는 여러 변수를 통하여 $L$에 영향을 준다.  

마찬가지로 연쇄법칙을 한 번 적용해 다음을 구한다.  
$$
\frac{\partial L}{\partial s_i}=\frac{\partial L}{\partial u_1}\frac{\partial u_1}{\partial s_i}+\frac{\partial L}{\partial u_2}\frac{\partial u_2}{\partial s_i}+\frac{\partial L}{\partial u_3}\frac{\partial u_3}{\partial s_i}+\frac{\partial L}{\partial u_4}\frac{\partial u_4}{\partial s_i}\ \text{for each}\  i
$$

그 다음 $w$의 관점에선 $L$을 완전함수 $L(s_1, s_2, s_3)$로 볼 수 있다.
- 이유: $s_1, s_2, s_3$의 값만 정해지면 $L$의 값이 구해진다.
- 따라서 다음의 연쇄법칙을 적용할 수 있다.
$$
\frac{\partial L}{\partial w} = \frac{\partial L}{\partial s_1}\frac{\partial s_1}{\partial w} + \frac{\partial L}{\partial s_2}\frac{\partial s_2}{\partial w} + \frac{\partial L}{\partial s_3}\frac{\partial s_3}{\partial w}
$$

각 $i$에 대해 $\frac{\partial s_i}{\partial w}$를 구한다.  
그러면 결국 $\frac{\partial L}{\partial s_i}$와 $\frac{\partial s_i}{\partial w}$를 위 식에 대입해 $\frac{\partial L}{\partial w}$를 구할 수 있게 된다.

#### 연쇄법칙의 간략화

$w$는 $s$층을 통해 $L$에 영향을 준다. 따라서 $w$의 관점에서 $L$을 $s_i$들의 완전함수 $L(s_1, s_2, s_3)$로 본다.

그런데 $w$는 $s$층의 변수 $s_2$에게만 영향을 준다.
- 즉 $s_1, s_3$는 $w$와 관련이 없다. $w$의 변화에 따라 전혀 변하지 않으므로 $w$의 관점에선 상수이고, 미분계수 또한 0으로 나오게 된다.
- 결국 $\frac{\partial s_i}{\partial w}=0,\ \text{for}\ i=1,3$

이에 따라 chain-rule은 다음처럼 간략화된다.

$$

\frac{\partial L}{\partial w} = \xcancel{\frac{\partial L}{\partial s_1}\frac{\partial s_1}{\partial w}} + \frac{\partial L}{\partial s_2}\frac{\partial s_2}{\partial w} + \xcancel{\frac{\partial L}{\partial s_3}\frac{\partial s_3}{\partial w}}
$$
$$
=\frac{\partial L}{\partial s_2}\frac{\partial s_2}{\partial w}
$$

### 마지막 층 뉴런들의 활성화 함수

마지막 층의 활성화 함수로 선형 함수를 사용한다.
- 이는 identity 함수(항등함수)라고도 한다. $a^T(x)=x$
- $s$를 아무 함수로 통과하지 않고 바로 출력한다.
- 마지막 $T$층을 제외한 다른 모든 층들엔 비선형 함수들 중에서 활성화 함수를 선택해야 한다.
    - $sigmoid, tanh, relu,...$

#### Softmax: 확률 구하기

$v_i = f_i^T$라고 하자. $f_i^T$는 마지막 T층에 있는 뉴런 $i$에서 나온 출력 신호이다.

각 클래스 $i$에 대해, $y_i$는 마지막 층의 뉴런들에서 온 출력신호를 입력받는 소프트맥스 함수를 사용해 계산된다.
$$
y_i = \frac{e^{v_i}}{\sum_{k=1}^{m_L}e^{v_k}}
$$

여기서:  
분모는 정규화 상수  
$y_i$는 $i$번째 범주의 레이블에 대한 확률이다.

### 손실의 정의

손실 또는 비용은 최소회되어야 할 것이 되도록 정의되어야 한다.

훈련 예제 $(X,d)$에 대해, 타겟 벡터 $Z$는 아래와 같이 레이블 $d$에서 만들어진다.
- 분류기 개발에서, $d$는 입력 $X$에 대해 요구되는 클래스의 인덱스이다.
- 타겟 벡터 $Z=(z_1, ..., z_{n_c})$는 아래와 같이 레이블 $d$에서 만들어진다.

$$
z_i = \begin{cases}
1\ if\ i=d\\
0\ otherwise
\end{cases}
$$

모델은 소프트맥스함수의 출력에 따라 $Y=(y_1, ..., y_{n_c})$를 출력한다.  
그러면 손실 교차 엔트로피(cross entropy)는 아래와 같이 정의된다:
$$
L=\sum_{i=1}^{n_c}z_i\ln(\frac{1}{y_i})=\ln(1/y_d)=-\ln(y_d)
$$

이 손실은 하나의 훈련 예제에 의한 손실량임을 주목하라.

목표는 $y_d$를 최대한 $1.0$에 가깝게 최대화하는 것이다.  
여기에 역수를 취하면 반대로 $1/y_d$를 최소화하는 문제가 된다.

# NN 모델의 Backpropagation 알고리즘: NN에서 gradient 계산의 표준 이론

신경망에서 손실 $L$의 모든 파라미터들에 대한 그레디언트 $\nabla L$을 구하는 알고리즘이다.  
이 결과는 경사하강법에 의해 파라미터의 갱신에 사용된다.  
딥러닝, 즉 현대 AI의 밑바탕에 되는 유일무이한 알고리즘이다.

NN는 다층구조로서 내부에는 신호값을 가지는 많은 변수가 존재한다.
- 파라미터 $w$에 대한 그레디언트를 구하기 위해선 $w$와 손실 $L$ 사이에 개입되는 모든 변수들의 그레디언트도 함께 구할 수 밖에 없다.

아이디어: chain-rule의 반복 적용  
위 층의 파라미터들의 그레디언트가 낮은 층과 연관된 파라미터들보다 먼저 계산되어야 한다.

모델의 파라미터:
- $W_{i, j}^n$: $(n-1)$층에서 $n$층으로 가는 연결의 가중치. 
    - $n$의 범위: $1 \sim T$
    - $i$의 범위: $1\sim(m_{n-1})$
    - $j$의 범위: $1\sim m_n$
- $b_j^n$: $n$층 뉴런들의 편향.

모든 뉴런에 새 변수 $\delta^n_i$를 추가한다.
- $\delta_i^n = \frac{\partial L}{\partial s_i^n}$

먼저 $\frac{\partial L}{\partial y_i}$를 구하자.
- $L=-\ln(y_d)$
- $\text{Thus}\ \frac{\partial L}{\partial y_i}\begin{cases}
0\ if\ i \neq d \\
-\frac{1}{y_d}\ if\ i == d
\end{cases}
$

그런 다음 $\frac{\partial L}{\partial f_i^T}$를 구하자.
- $\frac{\partial L}{\partial f_i^T}=\sum_{k=1}^{m_T}\frac{\partial L}{\partial y_k}\frac{\partial y_k}{\partial f_i^T}$
    - $\frac{\partial L}{\partial y_d}\frac{\partial y_d}{\partial f_i^T}$
    - $-\frac{1}{y_d}\frac{\partial y_d}{\partial f_i^T}\ \forall\ i=1, ..., m_T$

이제 $\frac{\partial y_d}{\partial f_i^T}$를 구하자.

Case 1: $i \neq d$

$$
\frac{\partial y_d}{\partial f_i^T} = \frac{\partial}{\partial f_i^T} \left( \frac{e^{f_d^T}}{\sum_{k=1}^{m_T} e^{f_k}} \right) 
= -e^{f_d} \times \frac{e^{f_i^T}}{\left( \sum_{k=1}^{m_T} e^{f_k} \right)^2}
$$

따라서,

$$
\frac{\partial L}{\partial f_i^T} = -y_d \frac{e^{f_i^T}}{\sum_{k=1}^{m_T} e^{f_k^T}} = \frac{e^{f_i^T}}{\sum_{k=1}^{m_T} e^{f_k^T}} \quad \text{(eq 2)}
$$

Case 2: $i = d$

$$
\frac{\partial y_d}{\partial f_d^T} = \frac{\partial}{\partial f_d^T} \left( \frac{e^{f_d^T}}{\sum_{k=1}^{m_T} e^{f_k^T}} \right) 
= \frac{e^{f_d^T}}{\sum_{k=1}^{m_T} e^{f_k^T}} - \frac{e^{f_d^T}}{\left( \sum_{k=1}^{m_T} e^{f_k^T} \right)^2} \times e^{f_d^T}
$$

따라서,

$$
\frac{\partial L}{\partial f_d^T} = -1 + \frac{e^{f_d^T}}{\sum_{k=1}^{m_T} e^{f_k^T}}
$$

종합:

$$
\frac{\partial L}{\partial f_i^T} = \begin{cases} 
\frac{e^{f_i^T}}{\sum_{k=1}^{m_T} e^{f_k^T}} & \text{if } i \neq d \\
\frac{e^{f_d^T}}{\sum_{k=1}^{m_T} e^{f_k^T}} - 1 & \text{if } i = d
\end{cases}
$$


몫의 미분법:

$$
\frac{d}{dx} \left( \frac{u(x)}{v(x)} \right) = \frac{v(x) \cdot u'(x) - u(x) \cdot v'(x)}{(v(x))^2}
$$

소프트맥스:

$$
y_d = \frac{e^{f_d^T}}{\sum_{k=1}^{m} e^{f_k^T}}
$$

여기서:

- $ u(f) = e^{f_d^T} $는 분자.
- $ v(f) = \sum_{k=1}^{m} e^{f_k^T} $는 분모.

몫의 미분법 적용:

$$
\frac{\partial y_d}{\partial f_i^T} = \frac{\left( v(f) \cdot \frac{\partial}{\partial f_i^T} u(f) - u(f) \cdot \frac{\partial}{\partial f_i^T} v(f) \right)}{(v(f))^2}
$$

- 분자의 미분 $ \frac{\partial}{\partial f_i^T} u(f) $:
    - 분자 $ u(f) = e^{f_d^T} $는 $ f_d^T $에만 의존, $ i \neq d $인 경우에는 0

$$
\frac{\partial}{\partial f_i^T} e^{f_d} = 0 \quad \text{(if } i \neq d \text{)}
$$

- 분모의 미분 $ \frac{\partial}{\partial f_i^T} v(f) $:
    - 분모 $ v(f) = \sum_{k=1}^{m} e^{f_k^T} $는 각 $ f_k^T $에 의한 지수 함수들의 합이다.

$$
\frac{\partial}{\partial f_i^7} \sum_{k=1}^{m} e^{f_k^T} = e^{f_i^T}
$$

정리:  
몫의 미분법 공식을 적용.

$$
\frac{\partial y_d}{\partial f_i^T} = \frac{ \left( \sum_{k=1}^{m} e^{f_k^T} \cdot 0 - e^{f_d^T} \cdot e^{f_i^T} \right)}{\left( \sum_{k=1}^{m} e^{f_k^T} \right)^2}
$$

$$
\frac{\partial y_d}{\partial f_i} = \frac{- e^{f_d} \cdot e^{f_i}}{\left( \sum_{k=1}^{m} e^{f_k} \right)^2}
$$

## 마지막 층에서의 역전파

$T$층의 활성화 함수는 선형함수이다.
$$
f_I^T = s_i^T
$$

$\frac{\partial L}{\partial s_i^T}$를 계산하기 위해 각 $s_i^T$가 그 뉴런으로부터 $f_i^T$에만 값을 전달한다는 점을 이용해 연쇄법칙을 단순화할 수 있다.

$$
\frac{\partial L}{\partial s_i^T}=\frac{\partial L(f_1^T, ..., f_{m_T}^T)}{\partial s_i^T}
$$
$$
=\frac{\partial L}{\partial f_i^T}\frac{\partial f_i^T}{\partial s_i^T}\ \text{(simplified chain rule)}
$$
$$
=\frac{\partial L}{\partial f_i^T},\ \text{(since }f_i^T=s_i^T)
$$
$$
y_i - 1_{i==d}
$$
$$
\text{Thus }\delta_i^T = \frac{\partial L}{\partial s_i^T}
$$
$$
=y_i - 1_{i==d}
$$

## 최종층의 아래층들에서의 역전파

$\delta$ 변수를 $n<T$인 $n$층의 변수라고 하자. (비 최종층 뉴런의 $\delta$)

단, $n+1$층의 $\delta_j^{n+1}=\frac{\partial L}{\partial s_i^{n+1}}\ \forall j$는 이미 구했다고 가정하자.

먼저 $\frac{\partial L}{\partial f_i^n}$를 구하자.

$$
\frac{\partial L}{\partial f_j^n} = \frac{\partial L(s_1^{n+1},...,s_{m_{n+1}}^{n+1})}{\partial f_j^n}
$$
$$
= \sum_{k=1}^{m_{n+1}}\frac{\partial L}{\partial s_k^{n+1}}\frac{\partial s_k^{n+1}}{\partial f_j^n}\ \text{(by chain rule)}
$$
$$
\sum_{k=1}^{m_{n+1}}\delta_k^{n+1}\frac{\partial s_k^{n+1}}{\partial f_j^n}
$$
$$
\text{since}\ s_k^{n+1}=(\sum_{r=1}^{m_n}w_{r,k}^{n+1})+b_k^{n+1}
$$
$$
\frac{\partial s_k^{n+1}}{\partial f_j^n} = w_{j,k}^{n+1}
$$

결국,

$$
\frac{\partial L}{\partial s_j^n} = \frac{\partial L(f_1^n, ..., f_{m_n}^n)}{\partial s_j^n}
$$
$$
=\frac{\partial L}{\partial f_j^n}\frac{\partial f_j^n}{\partial s_j^n}
$$

그런데 $\frac{\partial f_j^n}{\partial s_j^n} = a'(s^n_j)$, since $f^n_j = a(s^n_j)$

따라서:
$$
\delta^n_j \equiv \frac{\partial L}{\partial s_j^n} = \left(\sum^{m_{n+1}}_{k=1}\delta^{n+1}_k w^{n+1}_{j,k} \right)a'(s^n_j)
$$

$n+1$층의 $\delta$를 이용하여 $n$ 층의 $\delta$를 구한 기법을 그대로 $n-1$층 $\delta$를 구하는 데 사용할 수 있다. 이러한 방식을 $T$층에서 시작해 반복적용하여 모든 층에 대한 $\delta^n_j \equiv \frac{\partial L}{\partial s_j^n}$를 구한다.

윗층에서 구해진 $\delta$들이 아래층의 $\delta$를 구하는 데 연쇄적으로 사용되는 구조이기 때문에 역전파(backpropagation)이라고 한다.

## 파라미터의 그레디언트 구하기

### 편향 $b^n_j$의 그레디언트

$$
\frac{\partial L}{\partial b^n_j}=\frac{\partial L}{\partial s^n_j}\frac{\partial s^n_j}{\partial b^n_j}
$$

$$
s_n^j = b^n_j + \sum^{m_{n+1}}_{k=1}[W^n_{k,j} \cdot f^{n-1}_k] \ \forall\ 1\le n\le L\ \text{and all } 1 \le j \le m_n
$$

$$
\frac{\partial s^n_j}{\partial b^n_j} = \frac{\partial \left[ b^n_j + \sum^{m_{n+1}}_{k=1}[W^n_{k,j} \cdot f^{n-1}_k] \right]}{\partial b^n_j} = 1
$$

이유:
- $f^{n-1}_k$는 $b^n_j$와는 무관하다.

따라서:
$$
\frac{\partial L}{\partial b^n_j}=\frac{\partial L}{\partial s^n_j}
$$
$$
=\delta^n_j\ \forall\ n,\ 1\le n \le T\ and\ all\ j,\ 1\le j \le m_n
$$

### 가중치의 그레디언트 계산

임의의 weight parameter $W^n_{r,j}$의 gradient를 구하자.

$$
\frac{\partial L }{\partial W^n_{r,j}} = \frac{\partial L(s^n_j) }{\partial W^n_{r,j}}
$$
$$
= \frac{\partial L }{\partial s^n_j}\frac{\partial s^n_j}{\partial W^n_{r,j}}
$$

$$
s^n_j = b^n_j +\sum^{m_{n-1}}_{i=1}[W^n_{i,j} \cdot f^{n-1}_i]
$$

$$
\frac{\partial s^n_j}{\partial W^n_{r,j}} = f^{n-1}_r
$$

따라서:
$$
\frac{\partial L }{\partial W^n_{r,j}} = \delta^n_j f^{n-1}_r\ \forall\ n,j,r
$$

## 신경망의 역전파 알고리즘 종합

0. 먼저 주어진 예제의 입력에 대해 순전파(forward pass)를 수행한다. 이것으로 $L$이 결정된다.

1. 먼저 최종층($T$층)의 모든 뉴런들의 $\delta$를 구한다: $\delta^T_i = \frac{e^{f^T_i}}{\sum^{m_T}+{k=1}e^{f^T_k}}-1_{i=d}$
- 이 층의 모든 뉴런들마다 다음 2단계를 수행한다.
    - $b^T_j$의 그레디언트를 구한다: $\frac{\partial L}{\partial b^T_j} = \delta^T_j$
    - $W^T_{r,j}$의 그레디언트를 구한다: $\frac{\partial L }{\partial W^T_{r,j}} = \delta^n_j f^{T-1}_r\ \forall\ r$

2. for ($n=T-1$;$n > 0$; $n--$) {  
    $n$층의 모든 뉴런(모든 $j$)에 대해 다음 3단계를 수행한다.  
    $\delta^n_k$를 구한다: $\delta^{n}_{j} \equiv \frac{\partial L}{\partial s_j^n} = \left(\sum^{m_{n+1}}_{k=1}\delta^{n+1}_k w^{n+1}_{j,k} \right)a'(s^n_j)$  
    $b^n_j$의 그레디언트를 구한다: $\frac{\partial L}{\partial b^n_j} = \delta^n_j$  
    $W^n_{r,j}$의 그레디언트를 구한다: $\frac{\partial L }{\partial W^n_{r,j}} = \delta^n_j f^{n-1}_r\ \forall\ r$ \
}

위의 단계 1과 2를 __backward pass__ 라고 부른다.
